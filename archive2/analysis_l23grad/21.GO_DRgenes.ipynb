```python
import numpy as np
import pandas as pd
```

```python
f1 = '/data/qlyu/v1/data/cheng21_cell_scrna/res/L23-ABC-genes-n288-n286unq-annot.csv' 
f2 = '/data/qlyu/v1/results/P28NR_ITgrad_genes_n749_n338unq_230420.csv'

df1 = pd.read_csv(f1)
df2 = pd.read_csv(f2)


df1['anno_L23'] = df1['P17on']
df1 = df1.set_index('gene')
df2 = df2.groupby('gene', sort=False).sum()
df2['anno_IT'] = df2['anno']
```
